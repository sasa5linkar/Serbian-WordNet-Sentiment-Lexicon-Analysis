# Izrada Sentiment leksikona

Odve se  modeli trenirani u spyder-u koriste za izdradu sentiment lekiskona u vidu tablica
Modeli su snimljeni u ml_models folderu. Sripsi Wordent je u resources folderu


```python
import sys
sys.path.append('../')
from srpskiwordnet import SrbWordNetReader
from joblib import load
from SerbainTagger import SrbTreeTagger
import pandas as pd
```

```python

RES_DIR = "..\\resources\\"
MOD_DIR = "..\\ml_models\\"
```

```python
swordnet = SrbWordNetReader(RES_DIR,"wnsrp30.xml")
```

```python
POLARITY = ["POS", "NEG"]
TYPE = ["SVM", "Bern"]
ITERATION  =["0","2","4", "6"]
pom = list()

for p in POLARITY:
    for t in TYPE:
        for i in ITERATION:
            pom2 = dict()
            pom2["Iteracija"] =i
            pom2["Tip"] = t
            pom2["Polaritet"] = p
            pom2["Model"] = load (MOD_DIR + t + "-klas-" + p + "-" + i + ".joblit")
            pom.append(pom2)
estimators = pd.DataFrame(pom)


est_NEG = load (MOD_DIR + "Test-Bern-klas-NEG-0.joblit")
est_POS = load (MOD_DIR + "Test-Bern-klas-POS-0.joblit")
estimators
```

Ucitava se lematizator za preprocerinje. Ovo je wrapper za trettager obucen na Srpskom jeziku. Moze se zamentni bilo kojim stpskim lematizatorm. 

```python

tt = SrbTreeTagger()
tekst = ["veoma dobro delo", "putem nije bilo odstoja", "zla devojka"]
tekst = pd.DataFrame(tekst, columns=["Text"]).applymap(tt.lemmarizer)
tekst
```

Ovde pravimo prvu tablice. Tablice parvimo tako sto zamenjijemo u vrednosti koje vec postoje u wordnetu onim dobijenim već treniranim modelima. Prvo definisima dve pomocne funkcije. Prvu koja računa vrednost iz para modela treniranih pozivnim i negativim skupovima iste iteracije istog modele, and drugu koje nalayi vrednost od svih in datfrejma. 

Da bismo dobili vrednosti (0,0) od (1,1) a ostalo nepromenjeno korismo formoleu

 (a,b) ->  (a*(1-b), b*(1-a) 
 
 Napomena: zbog estimatora tekst je lista sa jednim String elementom. 


```python
def estimate_polarity(text, est_POS, est_NEG):
    a = est_POS.predict(text)[0]
    b = est_NEG.predict(text)[0]
    return ( a*(1-b), b*(1-a))
```

```python
filter1 = estimators["Iteracija"] == "2"
filter2 = estimators["Tip"] == "Bern"
filter3 = estimators["Polaritet"] =="NEG"
estimators.where(filter1 & filter2 & filter3).dropna().reset_index()["Model"][0]
```

Druga fumnkija od untetog teksta nd tablice modela nalazi srednju vrednost polariteta. 

```python
def estimate_polatiry_table (text, df):
    sentiment= list()
    for t in TYPE:
        filter2 = df["Tip"] == t
        for i in ITERATION:
            filter1 = df["Iteracija"] == i
            filter3 = df["Polaritet"] =="NEG"
            est_NEG = estimators.where(filter1 & filter2 & filter3).dropna().reset_index()["Model"][0]
            filter3 = df["Polaritet"] =="POS"
            est_POS = estimators.where(filter1 & filter2 & filter3).dropna().reset_index()["Model"][0]
            sentiment.append(estimate_polarity(text, est_POS, est_NEG))
    
    res = tuple([sum(ele) / len(sentiment) for ele in zip(*sentiment)])
    return res    
```

```python
for ime in swordnet._synset_dict:
    sinset = swordnet._synset_dict[ime]
    definicija = [tt.lemmarizer(sinset.definition())]
    if sinset.definition() is not None: 
        swordnet._synset_dict[ime]._sentiment = estimate_polatiry_table(definicija, estimators)
```

```python
sent_df =  swordnet.sentiment_df()
```

```python
def sentiment_anlyse_df(swn):
    syns_list = list()
    for sifra in swn._synset_dict:
        syn = swn._synset_dict[sifra]
        el = dict()
        el["ID"] = sifra
        el["POS"], el["NEG"] = syn._sentiment
        el["Lemme"] = ",".join(syn._lemma_names)
        el["Definicija"] = syn.definition()
        el["Vrsta"] = syn.POS() 
        syns_list.append(el)
    return  pd.DataFrame(syns_list)
```

```python
sent_an = sentiment_anlyse_df(swordnet)
```
